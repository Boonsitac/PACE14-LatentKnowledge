```python
# Loading the papers into the model
from sentence_transformers import SentenceTransformer

# Choose the model
# model = SentenceTransformer('all-MiniLM-L6-v2') # Fast and lightweight model
model = SentenceTransformer('all-mpnet-base-v2') # More powerful model
# model = SentenceTransformer('allenai-specter') # Model optimised for scientific papers

# Encode the "meaning" of the title + abstract into numerical form
# Papers with similar meanings will have vectors close together
X = model.encode(corpus)
```

```python
# Run HDBSCAN and DBSCAN
import hdbscan
from sklearn.cluster import DBSCAN

hdbscan_labels = hdbscan.HDBSCAN(min_cluster_size=2).fit_predict(X)
dbscan_labels = DBSCAN(eps=1.0, min_samples=2, metric='euclidean').fit_predict(X)
```

```python
# UMAP in 2D and 3D
# This transforms the semantic meaning vectors from the SentenceTransformer into 2D and 3D plots
# The transformation does its best to maintain the semantic meaning between similar papers

import umap

reducer_2d = umap.UMAP(n_components=2, random_state=42)
X_umap_2d = reducer_2d.fit_transform(X)

reducer_3d = umap.UMAP(n_components=3, random_state=42)
X_umap_3d = reducer_3d.fit_transform(X)
```

```python
# Plotting and Visualising the Algorithms
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_2d(embeddings, labels, title):
    plt.figure(figsize=(10, 7))
    scatter = plt.scatter(embeddings[:, 0], embeddings[:, 1], c=labels, cmap='tab10', s=70)
    for i, text in enumerate([p['title'][:25] + "..." for p in papers]):
        plt.annotate(f"{labels[i]}: {text}", (embeddings[i, 0], embeddings[i, 1]), fontsize=8)
    plt.colorbar(scatter, label='Cluster ID')
    plt.title(title)
    plt.xlabel("UMAP Dim 1")
    plt.ylabel("UMAP Dim 2")
    plt.grid(True)
    plt.show()

def plot_3d(embeddings, labels, title):
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    sc = ax.scatter(embeddings[:, 0], embeddings[:, 1], embeddings[:, 2], c=labels, cmap='tab10', s=60)
    for i, text in enumerate([p['title'][:25] + "..." for p in papers]):
        ax.text(embeddings[i, 0], embeddings[i, 1], embeddings[i, 2], f"{labels[i]}: {text}", fontsize=7)
    ax.set_title(title)
    ax.set_xlabel("UMAP Dim 1")
    ax.set_ylabel("UMAP Dim 2")
    ax.set_zlabel("UMAP Dim 3")
    fig.colorbar(sc, label='Cluster ID')
    plt.show()

# Visualise all combinations
plot_2d(X_umap_2d, hdbscan_labels, "2D UMAP Clustering (HDBSCAN)")
print("\n" * 2)

plot_2d(X_umap_2d, dbscan_labels, "2D UMAP Clustering (DBSCAN)")
print("\n" * 2)

plot_3d(X_umap_3d, hdbscan_labels, "3D UMAP Clustering (HDBSCAN)")
print("\n" * 2)

plot_3d(X_umap_3d, dbscan_labels, "3D UMAP Clustering (DBSCAN)")


```

```python
# Cluster summary table
import pandas as pd
pd.set_option('display.max_colwidth', None)

df = pd.DataFrame({
    'Title': [p['title'] for p in papers],
    'HDBSCAN Cluster': hdbscan_labels,
    'DBSCAN Cluster': dbscan_labels
})
display(df.sort_values(['HDBSCAN Cluster', 'DBSCAN Cluster']))
```

```python
# Using OPTICS Clustering (OPTICS: Ordering Points to Identify the Clustering Structure)
from sklearn.cluster import OPTICS

# Run OPTICS on sentence embeddings
optics = OPTICS(min_samples=2, xi=0.05, min_cluster_size=0.05)
optics_labels = optics.fit_predict(X)

# 2D UMAP plot for OPTICS
plot_2d(X_umap_2d, optics_labels, "2D UMAP Clustering (OPTICS)")
print("\n" * 2)

# 3D UMAP plot for OPTICS
plot_3d(X_umap_3d, optics_labels, "3D UMAP Clustering (OPTICS)")
print("\n" * 2)

df = pd.DataFrame({
    'Title': [p['title'] for p in papers],
    'HDBSCAN Cluster': hdbscan_labels,
    'DBSCAN Cluster': dbscan_labels,
    'OPTICS Cluster': optics_labels
})
display(df.sort_values(['OPTICS Cluster']))
```
